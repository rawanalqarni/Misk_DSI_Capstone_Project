```python
import selenium
from selenium import webdriver 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


path = "C:/Users/hmq4/OneDrive/Desktop/Data Science- Misk/chrome/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("https://www.goldenscent.com/en/c/arabic/new-in.html?action=cat&id=2584")
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import time

productlinks = driver.find_elements(By.CLASS_NAME, 'ais-InfiniteHits-item.col-xs-6.col-md-4.flex')
links = []
for i in productlinks:
    url = i.find_element(By.CLASS_NAME ,'block.no-underline.product-link').get_property('href')
    links.append(url)
print(len(links))
```

```python
import wget
import os 
details=[]   
for j in links:
    url = webdriver.Chrome(path)
    url.get(j)
    try:       
        name = url.find_element(By.XPATH , '//*[@id="pp-gallery-n-details-wrapper"]/div/div/div[2]/div/div[1]/div/div[1]/h1').text
        price = url.find_element(By.XPATH ,'//*[@id="pp-gallery-n-details-wrapper"]/div/div/div[2]/div/div[2]/div/div/div[1]/div/div[1]/span[1]').text
        description = url.find_element(By.XPATH, '//*[@id="section-description"]/div/div/div[2]/p').text        
    except:
        name = "none"
        price = "none"
        description = "none"
    try:
        rate = url.find_element(By.XPATH ,'//*[@id="pp-gallery-n-details-wrapper"]/div/div/div[2]/div/div[1]/div/div[1]/div/span/span').text
    except:
        rate = "none"
    
    try:
        rate_count = url.find_element(By.XPATH ,'//*[@id="rating-wrapper"]/div[1]/div/div[2]').text
    except:
        rate_count= "none"

    try:
        img_url = url.find_element(By.XPATH, '//*[@id="pp-gallery-n-details-wrapper"]/div/div/div[1]/div/div/div[2]/section/div[1]/div/div/div/img[3]').get_attribute('src')
        destination = "images/"+ img_url.split('/')[-1]
        if os.path.exists(destination) != True:
            wget.download(img_url,destination)
    except:
        img_url="none"
    
    try: 
        all_details=[]
        all = url.find_elements(By.XPATH , '//*[@id="section-info"]/div/div/div[2]/ul/li')
        for i in all:      
            title = i.find_element(By.CSS_SELECTOR, '.col-md-2').text
            if title == 'Brand':
                value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                all_details.append((title,value))
            else:
                value = "none"
            if title == 'Gender':
                value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                all_details.append((title,value))
            else:
                value = "none"
            if title == 'Product Type':
                value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                all_details.append((title,value))
            else:
                value = "none"
            if title == 'Character':
                value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                all_details.append((title,value))
            else:
                value = "none"
            if title == 'Fragrance Family':
                value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                all_details.append((title,value))
            else:
                value = "none"
            if title == 'Size':
                value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                all_details.append((title,value))
            else:
                value = "none"
            if title =='Year of Launch':
                    value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                    all_details.append((title,value))
            else:
                value = "none"
            if title =='Ingredients':
                    value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                    all_details.append((title,value))
            else:
                value = "none"
            if title =='Concentration':
                    value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                    all_details.append((title,value))
            else:
                value = "none"
            if title =='Top Notes':
                value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                all_details.append((title,value))
            else:
                value = "none"
            if title =='Middle Notes':
                value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                all_details.append((title,value))
            else:
                value = "none"
            if title =='Base Notes':
                value = i.find_element(By.CSS_SELECTOR, '.col-md-10').text
                all_details.append((title,value))
            else:
                value = "none"
            
    except:
        all = "none"
 
    details.append((name, price, description, rate,rate_count ,all_details, img_url))
```

```python
import pandas as pd 

Arabic_Perfume =pd.DataFrame(details,columns=['Name', 'Price','Description','Rate', 'Rating_count','Details','image'])
split_df = pd.DataFrame(Arabic_Perfume['Details'].tolist(), columns=['Brand', 'Gender', 'Product_Type', 'Character', 'Fragrance_Family','Product Color', 'Size', 'Year', 'Ingredients', 'Concentration', 'Top_note', 'Middle_note', 'Base_note'])
Arabic_Perfume.join(split_df)
Arabic_Perfume.to_csv('Arabic_Perfume.csv',index=False)
```
